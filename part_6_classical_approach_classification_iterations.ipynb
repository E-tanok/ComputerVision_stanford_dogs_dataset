```python
import time
import pickle
import pandas as pd
import numpy as np
```

```python
from context import pickles_path, datasources_path
```

```python
result_schema = ['n_races','n_clusters','classifier','cv_accuracy','final_accuracy','best_params']
```

```python
N_RACES = pickle.load(open(pickles_path+"N_FILTERED_RACES_last_execution.p", "rb" ))
```

```python
global_start_time = time.time()
```

<p>
        **<font color='#D2691E'size="4">I) Looping over the classification program parameters</font>**
</p>

```python
#Starting to iterate on the number of clusters : 
for N_CLUSTERS in [50,100,200,300,500]:
    execution_params_workflows_iteration = {'N_RACES':N_RACES,
                                            'N_CLUSTERS':N_CLUSTERS}
    #We dump our execution params :
    pickle.dump(execution_params_workflows_iteration,open(pickles_path+"execution_params_clusters_iteration.p", "wb"))
    %run -i 2_approche_classique_1_classification_KM.ipynb
    
    sub_results_df = pickle.load(open(pickles_path+"1_cluster_iteration_results_df.p", "rb" ))
    sub_results_df.to_csv(datasources_path+"results.csv", mode='a', header=False)
```

```python
print("--- %s seconds ---" % (time.time() - global_start_time))
```
