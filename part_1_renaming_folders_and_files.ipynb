```python
from os import listdir
import numpy as np
import os
import re
```

```python
from context import datasources_path
```

```python
races_path = datasources_path+"Images\\"
races_path
```

```python
from functions import class_my_files
```

<p>
        **<font color='#D2691E'size="4">I) Renaming the directories</font>**
</p>

```python
L_directories = listdir(races_path)
L_directories
```

```python
# renaming the folders:
for directory in L_directories:
    clean_name = re.sub('^n\d{8}[_-]','',directory)
    clean_name = re.sub('-','_',clean_name)
    clean_name = str.lower(clean_name)
    os.rename(races_path+directory,races_path+clean_name)
```

```python
L_directories = listdir(races_path)
L_directories
```

<p>
        **<font color='#D2691E'size="4">II) Renaming the pictures</font>**
</p>

```python
dict_race_pictures = {}
for directory in L_directories:
    dict_extensions_files = class_my_files(races_path+directory)
    dict_race_pictures[directory] = dict_extensions_files['jpg']
dict_race_pictures
```

```python
L_races = []
for race in dict_race_pictures.keys():
    L_races.append(race)
L_races
```

```python
for race in L_races:
    for picture_name in dict_race_pictures[race]:
        picture_path = races_path+race+"\\"+picture_name
        clean_picture_path = re.sub('n\d{8}[_-]',race+"_",picture_path)
        os.rename(picture_path,clean_picture_path)
```

```python
dict_race_pictures = {}
for directory in L_directories:
    dict_extensions_files = class_my_files(races_path+directory)
    dict_race_pictures[directory] = dict_extensions_files['jpg']
dict_race_pictures
```
