<p>
        **<font color='#D2691E'size="4">III) SIFT features agregation at the "Race" level</font>**
</p>

```python
#Analyze of occurencies per races :
```

```python
grouped = group_by_frequency(df_train,'race')
grouped.rename({'occurences':'feature_occurences'}, axis=1, inplace=True)
grouped
```

```python
plt.figure(figsize=(4,10))
plt.title('SIFT features occurences per races boxplot')
sns.boxplot(y="feature_occurences",
            palette=["orange"],
            data=grouped)
sns.despine(offset=10, trim=True)
#At the race level, the SIFT features distribution is now smoother, even if we have a "slightly strong" distance between the upper
#quantile and the median.
```

```python
groupped_per_occurencies = df_train[['race','picture_name']].copy()
groupped_per_occurencies.drop_duplicates(subset=['race','picture_name'], inplace=True)
groupped_per_occurencies = group_by_frequency(groupped_per_occurencies,'race')
groupped_per_occurencies = groupped_per_occurencies[['occurences']]
groupped_per_occurencies.rename({'occurences':'race_picture_occurences'}, axis=1, inplace=True)
groupped_per_occurencies
```

```python
plt.figure(figsize=(4,10))
plt.title('Number of pictures per races boxplot')
sns.boxplot(y="race_picture_occurences",
            palette=["red"],
            data=groupped_per_occurencies)
sns.despine(offset=10, trim=True)
#We can also observe that the number of pictures distribution per races is not balanced too : is there a relation between this
#imbalance and that of the number of SIFT features per races ?
```

```python
grouped = pd.merge(grouped, groupped_per_occurencies, how='left', left_index=True, right_index=True)
grouped
```

```python
grouped.plot.scatter('race_picture_occurences',
                     'feature_occurences',
                     logy =True,
                     figsize=(15,10),
                     title = "number of SIFT features depending on  the pictures occurences")
#We can observe a slight trend : the more we have input pictures, the more we get SIFT pictures
# If it's a true trend, we could balance this phenomenon by sampling exactly the same number of pictures for each race
```

```python
#Let's normalize the feature occurences per races by the number of pictures attached to races :
grouped['norm_feature_occurences'] = grouped['feature_occurences']/grouped['race_picture_occurences']
grouped
```

```python
plt.figure(figsize=(4,10))
plt.title('SIFT features occurences per races boxplot')
sns.boxplot(y="norm_feature_occurences",
            palette=["orange"],
            data=grouped)
sns.despine(offset=10, trim=True)
#The distribution is now smoother : we just have one outlier. If we look at the upper result, we can see that the "dhole" race
#burts the others in terms of SIFT features generated :
```

<p>
        **<font color='#D2691E'size="4">IV) SIFT features distributions per races</font>**
</p>

```python
L_races = np.sort(np.unique(df_train['race']))
L_races
```

```python
MIN_RANGE = 0
MAX_RANGE = 8
while MAX_RANGE<=128:
    L_feature_to_analyze = ['race']
    L_feature_to_analyze+= list(range(MIN_RANGE,MAX_RANGE))
    
    grouped = df_train[L_feature_to_analyze][df_train['race'].isin(L_races)].groupby('race').mean()
    grouped.plot(kind='bar',figsize=(15,4),title = "Average descriptors distribution per races : descriptors %s to %s"%(L_feature_to_analyze[1],L_feature_to_analyze[-1]))
    plt.show()
    MIN_RANGE+= 8
    MAX_RANGE+= 8
#In terms of average per race, descriptors distributions are similar : this is not a good thing because we need to differenciate
#each race in order to make the classification.
#We could think that this phenomenon is dued to the fact that a large part of the keypoints detected are not on the dogs themself
#If races are quite different, the environments where they are photographed can be close to each other : they contain noise and 
#this noise is reflected on all the descriptors

#A second asssumption could be that by construction SIFT descriptors are nothing that informations about the gradient of 
#light intensity around the feature itself. If we agregate all this informations at the race level we don't do anything else
# than agregating randomness. We can't detect patterns inrandomness so all distributions seams to be identical.
```

```python
MIN_RANGE = 0
MAX_RANGE = 8
while MAX_RANGE<=128:
    L_feature_to_analyze = ['race']
    L_feature_to_analyze+= list(range(MIN_RANGE,MAX_RANGE))
    
    grouped = df_train[L_feature_to_analyze][df_train['race'].isin(L_races)].groupby('race').sum()
    grouped.plot(kind='bar',figsize=(15,4),title = "Sum of descriptors distribution per races : descriptors %s to %s"%(L_feature_to_analyze[1],L_feature_to_analyze[-1]))
    plt.show()
    MIN_RANGE+= 8
    MAX_RANGE+= 8
#In terms of sum per race, descriptors distributions are not similar but proportional to each other :
#We can make the same assumption as before to explain this.
```
