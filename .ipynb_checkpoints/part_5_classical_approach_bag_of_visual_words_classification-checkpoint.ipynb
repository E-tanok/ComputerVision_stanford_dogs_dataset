```python
import pandas as pd
import numpy as np
import pickle
import time
```

```python
from context import datasources_path, pickles_path, temp_files_path
```

```python
from functions import values_to_col
```

```python
execution_params = pickle.load(open(pickles_path+"execution_params_clusters_iteration.p", "rb" ))
```

```python
N_RACES = execution_params['N_RACES']
N_CLUSTERS = execution_params['N_CLUSTERS']
print()
print()
print()
print('Iteration for params : N_RACES = %s | N_CLUSTERS = %s'%(N_RACES,N_CLUSTERS))
```

```python
print('Loading the train data ...')
df_model_train = pickle.load(open(pickles_path+"df_model_train_with_%s_clusters_%s_races.p"%(N_CLUSTERS,N_RACES), "rb" ))
df_model_train
```

```python
print('Loading the test data ...')
df_model_test = pickle.load(open(pickles_path+"df_model_test_with_%s_clusters_%s_races.p"%(N_CLUSTERS,N_RACES), "rb" ))
df_model_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

<p>
        **<font color='#D2691E'size="4">I) Train and test matrices and vectors building</font>**
</p>

```python
cluster_columns = ["cluster_%s"%i for i in range(0,N_CLUSTERS)]
```

```python
X_train = df_model_train[cluster_columns].values
X_test = df_model_test[cluster_columns].values

y_train = list(df_model_train['encoded_race'].values)
y_test = list(df_model_test['encoded_race'].values)
```

```python
from sklearn import model_selection
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
df_results = pd.DataFrame(columns = ['n_races','n_clusters','classifier','cv_accuracy','final_accuracy','best_params'])
```

```python
global_start_time = time.time()
```

```python
print('Starting the classification ...')
```

<p>
        **<font color='#D2691E'size="4">II) Building the classifiers with cross validation </font>**
</p>

<p>
        **<font color='#D2691E'size="4">II.1) Logistic regression</font>**
</p>

<p>
        **<font color='#D2691E'size="4">II.1.A) Logistic regression : cross validation</font>**
</p>

```python
print()
print('Logistic regression : cross validation ...')
```

```python
clf_lr  = model_selection.GridSearchCV(LogisticRegression(),
                                      {'penalty':['l1','l2'],
                                      'C':[0.1,0.5,1.0],
                                      'n_jobs':[4]},
                                      cv = 5,
                                      scoring = 'accuracy')

#C : Inverse of regularization strength; must be a positive float. Like in support vector machines, smaller values specify 
#stronger regularization.
```

```python
#We fit the grid search classifiers :
start_time = time.time()
clf_lr.fit(X_train,y_train)
print("--- %s seconds to fit the lr_count model---" % (time.time() - start_time))
```

```python
#Bests parameters are : 
best_params_lr= clf_lr.best_params_
print("best_params_lr_count : %s"%best_params_lr)
```

```python
#Resulting performances with each hyperparameters are : 
for mean, std, params in zip(clf_lr.cv_results_['mean_test_score'],
                            clf_lr.cv_results_['std_test_score'],
                            clf_lr.cv_results_['params']):
    print('logistic regression : \taccuracy = %0.3f (+/-%0.3f) for %s'%(mean,std*2,params))
```

```python
#Logistic regression : applying on the test data
```

```python
y_pred_lr = clf_lr.predict(X_test)
clf_lr_accuracy = metrics.accuracy_score(y_test,y_pred_lr)
print("final accuracy on test data for clf_lr model is : %s"%clf_lr_accuracy)
```

<p>
        **<font color='#D2691E'size="4">II.1.B) Logistic regression : final model</font>**
</p>

```python
print('Logistic regression : building the final model ...')
```

```python
start_time = time.time()
log_regr = LogisticRegression(penalty=best_params_lr['penalty'],
                                    C=best_params_lr['C'],
                                    n_jobs=4)

log_regr.fit(X_train,y_train)
y_pred_log_regr  = log_regr.predict(X_test)

log_regr_final_accuracy = metrics.accuracy_score(y_test,y_pred_log_regr )
print("Accuracy is : %s"%log_regr_final_accuracy)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df_results.columns
```

```python
tmp_results = {
    'n_races':N_RACES,
    'n_clusters':N_CLUSTERS,
    'classifier':'Logistic_regression',
    'cv_accuracy':clf_lr_accuracy,
    'final_accuracy':log_regr_final_accuracy,
    'best_params':best_params_lr
}
df_results = df_results.append(tmp_results, ignore_index=True)
df_results
```

<p>
        **<font color='#D2691E'size="4">II.2) Support vector classifier</font>**
</p>

<p>
        **<font color='#D2691E'size="4">II.2.A) Support vector classifier : cross validation</font>**
</p>

```python
print()
print()
print('Support Vector Classifier : cross validation ...')
```

```python
clf_svc  = model_selection.GridSearchCV(LinearSVC(),
                                      {'C':[0.1,0.5,1.0]},
                                       cv = 5,
                                      scoring = 'accuracy')
```

```python
#We fit the grid search classifiers :
start_time = time.time()
clf_svc.fit(X_train,y_train)
print("--- %s seconds to fit the svc model---" % (time.time() - start_time))
```

```python
#Bests parameters are : 
best_params_svc = clf_svc.best_params_
print("best_params_svc : %s"%best_params_svc)
```

```python
#Resulting performances with each hyperparameters are : 
for mean, std, params in zip(clf_svc.cv_results_['mean_test_score'],
                            clf_svc.cv_results_['std_test_score'],
                            clf_svc.cv_results_['params']):
    print('support vector classifier : \taccuracy = %0.3f (+/-%0.3f) for %s'%(mean,std*2,params))
```

```python
#Support Vector Classifier : applying on the test data
```

```python
y_pred_svc = clf_svc.predict(X_test)
clf_svc_accuracy = metrics.accuracy_score(y_test,y_pred_svc)
print("final accuracy on test data for clf_svc model is : %s"%clf_svc_accuracy)
```

```python
clf_lr_accuracy = metrics.accuracy_score(y_test,y_pred_lr)
print("final accuracy on test data for clf_lr model is : %s"%clf_lr_accuracy)
```

<p>
        **<font color='#D2691E'size="4">II.2.B) Support vector classifier : final model</font>**
</p>

```python
print('Support Vector Classifier : building the final model ...')
```

```python
start_time = time.time()
svc = LinearSVC(C = best_params_svc['C'])

svc.fit(X_train,y_train)
y_pred_svc = svc.predict(X_test)

svc_final_accuracy = metrics.accuracy_score(y_test,y_pred_svc)
print("Accuracy is : %s"%svc_final_accuracy)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
tmp_results = {
    'n_races':N_RACES,
    'n_clusters':N_CLUSTERS,
    'classifier':'Linear_svc',
    'cv_accuracy':clf_svc_accuracy,
    'final_accuracy':svc_final_accuracy,
    'best_params':best_params_svc
}

df_results= df_results.append(tmp_results, ignore_index=True)
df_results
```

<p>
        **<font color='#D2691E'size="4">II.3) Random forest</font>**
</p>

<p>
        **<font color='#D2691E'size="4">II.3.A) Random forest : cross validation</font>**
</p>

```python
print()
print()
print('Random Forest : cross validation ...')
```

```python
clf_rf  = model_selection.GridSearchCV(RandomForestClassifier(),
                                      {'n_estimators':[1000],
                                      'max_features':[2,5,10],
                                      'n_jobs':[4]},
                                      cv = 5,
                                      scoring = 'accuracy')
```

```python
#We fit the grid search classifiers :
start_time = time.time()
clf_rf.fit(X_train,y_train)
print("--- %s seconds to fit the random forest model---" % (time.time() - start_time))
```

```python
#Bests parameters are : 
best_params_rf = clf_rf.best_params_
print("best_params_rf : %s"%best_params_rf)
```

```python
#Resulting performances with each hyperparameters are : 
for mean, std, params in zip(clf_rf.cv_results_['mean_test_score'],
                            clf_rf.cv_results_['std_test_score'],
                            clf_rf.cv_results_['params']):
    print('random forest : \taccuracy = %0.3f (+/-%0.3f) for %s'%(mean,std*2,params))
```

```python
#Random Forest : applying on the test data
```

```python
y_pred_rf = clf_rf.predict(X_test)
clf_rf_accuracy = metrics.accuracy_score(y_test,y_pred_rf)
print("final accuracy on test data for clf_rf model is : %s"%clf_rf_accuracy)
```

<p>
        **<font color='#D2691E'size="4">II.3.B) Random forest : final model</font>**
</p>

```python
print('Random Forest : building the final model ...')
```

```python
start_time = time.time()
r_forest = RandomForestClassifier(n_estimators=best_params_rf['n_estimators'],
                                         max_features=best_params_rf['max_features'])

r_forest.fit(X_train,y_train)
y_pred_r_forest  = r_forest.predict(X_test) 

r_forest_final_accuracy = metrics.accuracy_score(y_test,y_pred_r_forest )
print("Accuracy is : %s"%r_forest_final_accuracy)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
tmp_results = {
    'n_races':N_RACES,
    'n_clusters':N_CLUSTERS,
    'classifier':'Random_forest',
    'cv_accuracy':clf_rf_accuracy,
    'final_accuracy':r_forest_final_accuracy,
    'best_params':best_params_rf
}

df_results= df_results.append(tmp_results, ignore_index=True)
```

```python
print()
print()
print()
print("--- %s seconds for this iteration---" % (time.time() - global_start_time))
```
